```python
# importing the csv file
import pandas as pd
data=pd.read_csv('breast-cancer-wisconsin-data_data.csv')
```

```python
# counting the number of malignant and benign cancer cases
target_counts = data['diagnosis'].value_counts()
print("Counts of Malignant (1) and Benign (0) cases:\n", target_counts)
```

```python
#plotting a heatmap for the dataset
correlation_matrix = data.corr()
plt.figure(figsize=(30, 30))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

```python
plt.bar(data[data['diagnosis']=='M']['diagnosis'],data[data['diagnosis']=='M']['radius_mean'],color='blue',label='Malignant radius')
plt.bar(data[data['diagnosis']=='B']['diagnosis'],data[data['diagnosis']=='B']['radius_mean'],color='green',label='Benign radius')
plt.legend()
```

```python
sns.histplot(data[data['diagnosis']=='M']['radius_mean'],color='blue',label='Malignant radius')
sns.histplot(data[data['diagnosis']=='B']['radius_mean'],color='green',label='Benign radius')
plt.legend()
```

```python
#taking x as input and y as the target variable
x=data.drop(columns='diagnosis')
y=data['diagnosis']
```

```python
x=x.fillna(0)
```

```python
x.isna().sum()
```

```python
x.head()
```

```python
y
```

**Model Building and Prediction**

```python
#splitting data into 80% train and 20% test
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=1)
```

```python
x_train.shape,y_train.shape
```

```python
x_test.shape,y_test.shape
```

Using different models to check for the best accuracy

**GBDT**

```python
from sklearn.ensemble import GradientBoostingClassifier
```

```python
gbdt=GradientBoostingClassifier(n_estimators=200,max_depth=3)
```

```python
gbdt.fit(x_train,y_train)
```

```python
gbdt.score(x_test,y_test)
```

**Decision Tree**

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model=DecisionTreeClassifier(criterion='entropy',max_depth=7)
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
model.get_depth()
```

**K-Nearest Neighbours**

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
model=KNeighborsClassifier(n_neighbors=5)
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

**Logistic Regression**

```python
from sklearn.linear_model import LogisticRegression
```

```python
model=LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

**Naive Bayes**

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
model=MultinomialNB()
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```
